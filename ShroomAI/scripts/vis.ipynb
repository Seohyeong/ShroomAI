```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import random
from collections import Counter
```

```python
df = pd.read_csv('../dataset/train_df.txt',  sep = '\t', dtype=str)
```

```python
as_df = df[df['phylum'] == 'Ascomycota']
Counter(as_df['class'])
```

```python
df = df[df['phylum'] == 'Ascomycota']

# Get value counts for 'class' column
class_counts = df['class'].value_counts()

# Define pastel color palette
colors = sns.color_palette('pastel')[0:len(class_counts)]

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, colors=colors, autopct=lambda p: '{:.0f}'.format(p * sum(class_counts) / 100), startangle=140)
plt.title('Count of Classes')
plt.show()
```

```python
img_dir = '/Users/seohyeong/Projects/ShroomAI/ShroomAI/dataset/images_100_combined/test'
label_dirs = [os.path.join(img_dir, label) for label in os.listdir(img_dir)]
img_paths = []
for label_dir in label_dirs:
    items = os.listdir(label_dir)
    for item in items:
        img_paths.append(os.path.join(label_dir, item))

random.sample(img_paths, 8)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Example image paths and corresponding labels
image_paths = random.sample(img_paths, 16)
print(image_paths)
labels = [img_path.split('/')[-2] for img_path in image_paths]

# Load images using matplotlib.image.imread into a list of numpy arrays
images = [mpimg.imread(img_path) for img_path in image_paths]

# Create a figure and axis with matplotlib
fig, axes = plt.subplots(4, 4, figsize=(12, 12))

# Iterate over the images and plot them in the grid with labels
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i])
    ax.axis('off')  # Turn off axis labels
    ax.set_title(labels[i], fontsize=13)  # Set title as the label

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
